# Executive Summary — ASRS Narrative Intelligence Pipeline

This notebook provides a high-level, narrative overview of the ASRS
(“Aviation Safety Reporting System”) text analytics pipeline developed
across Notebooks 01–09. The goal of this project is to transform
16,535 pilot and controller safety narratives into structured,
interpretable insights using modern NLP, clustering, HFACS analysis,
and risk scoring.

This summary highlights the major components, methods, and findings
from the full workflow.

## Why Analyze ASRS Narratives?

ASRS reports contain rich, unstructured descriptions of operational
issues, human-factor breakdowns, and near-miss events. While structured
fields capture metadata, the narrative often reveals the *real* root
cause.

This project builds a full applied-AI pipeline to:

- extract meaning from narrative text
- identify systemic human-factor contributors
- cluster incident themes
- score risk signals
- visualize operational patterns
- connect HFACS categories to cluster-level behavior

The result is a scalable, interpretable safety intelligence system.

## Major Findings

### 1. Distinct Narrative Themes
Clustering reveals clear operational themes such as:
- communication issues
- procedural deviations
- ATC misunderstandings
- aircraft configuration errors
- environmental factors

### 2. Strong HFACS Signals
HFACS flags show concentrated patterns:
- Unsafe Acts dominate several clusters
- Preconditions appear in weather/ATC clusters
- Unsafe Supervision emerges in procedural clusters
- Organizational Influences appear in training-related clusters

### 3. Risk Concentration
Risk scores highlight clusters with:
- repeated human-factor breakdowns
- multi-factor causal chains
- high-density anomaly patterns

### 4. Systemic Contributors
Narratives reveal recurring issues in:
- communication protocols
- workload management
- situational awareness
- procedural adherence

## Methods Summary

### Text Representation
TF-IDF vectorization of 16k+ narratives.

### Clustering
KMeans clustering with optimized k.

### Dimensionality Reduction
UMAP and t-SNE for 2D visualization.

### HFACS Extraction
Binary flags for:
- Unsafe Acts
- Preconditions
- Unsafe Supervision
- Organizational Influences

### Risk Scoring
Rule-based and signal-based scoring using:
- causal chains
- contributing factors
- human-factor indicators

## Final Takeaways

This pipeline demonstrates how unstructured aviation safety narratives
can be transformed into actionable intelligence. By combining NLP,
clustering, HFACS analysis, and risk scoring, we reveal patterns that
support:

- proactive safety management
- training improvements
- operational risk mitigation
- systemic issue identification

The workflow is fully reproducible, scalable, and interpretable.
